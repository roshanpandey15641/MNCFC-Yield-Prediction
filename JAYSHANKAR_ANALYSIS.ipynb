```python
data_total = pd.read_csv('Jayashankar_CCE_Extract_Two_Stage_250m.csv')
data = data_total.copy()
```

```python
corr = data.corr()
```

```python
corr.to_excel("Jayashankar_Corr.xlsx")
```

```python
corr
```

```python
data.drop(columns=['Long','State','District','Block_Mand','Village','CCE_Date','Target_Cro',
                   'Geo_Coordi','CCE_Plot_S','QC_Status','Lat','S_No_','USIN','Wet_Weight','Dry_Weight'],axis=1,inplace=True)
```

```python
data['Avg_max'] = (data.Jaya_Shankar_max_GS_250+data.Jaya_Shankar_max_VS_250)/2
data['Avg_min'] = (data.Jaya_Shankar_min_GS_250+data.Jaya_Shankar_min_VS_250)/2
data['Sum_rf'] = (data.Jaya_Shankar_rf_GS_250+data.Jaya_Shankar_rf_VS_250)
```

```python
### Thresholding is applied for outlier treatment
low_thresh = np.mean(data.Yield_kg_h)-2*np.std(data.Yield_kg_h)
upp_thresh = np.mean(data.Yield_kg_h)+2*np.std(data.Yield_kg_h)
```

```python
low_thresh
```

```python
upp_thresh
```

```python
selected_data = data[np.logical_and(data.Yield_kg_h > low_thresh, data.Yield_kg_h < upp_thresh)]
```

```python
X = selected_data.drop(['Yield_kg_h'],axis=1)
Y = selected_data.Yield_kg_h
```

```python
X.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 19, test_size=0.1)
rf = RandomForestRegressor(n_estimators=35, max_depth=9, min_samples_split=2, random_state=7)
rf.fit(X_train, Y_train)
x_pred = rf.predict(X_train)
y_pred = rf.predict(X_test)
print('Train score:', r2_score(x_pred, Y_train))
print('Test score:',r2_score(Y_test, y_pred))
devs = ((y_pred-Y_test)/Y_test)*100
print(np.mean(np.abs(devs)))
```

```python
import xgboost as xgb
```

```python
xg_reg = xgb.XGBRegressor(objective ='reg:squarederror', colsample_bytree = 0.4, learning_rate = .25,
                max_depth = 7, alpha = 10, n_estimators = 30)
xg_reg.fit(X_train,Y_train)

x_pred_xgb = xg_reg.predict(X_train)
y_pred_xgb = xg_reg.predict(X_test)
print("Train score: ", r2_score(x_pred_xgb,Y_train))
print("Test score: ", r2_score(Y_test, y_pred_xgb))
devs = ((y_pred_xgb-Y_test)/Y_test)*100
print(np.mean(np.abs(devs)))
```
