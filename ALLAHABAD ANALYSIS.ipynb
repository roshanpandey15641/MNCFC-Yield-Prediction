```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
plt.rcParams['font.size']=14
plt.rcParams['axes.titlepad']= 8
plt.rcParams['axes.titlesize']= 'medium'
plt.rcParams['axes.grid']=True
plt.rcParams['figure.figsize'] = (5,5)
plt.rcParams['axes.facecolor'] = 'white'
```

```python
data_total = pd.read_csv('30M_Train_data/Allahabad_CCE_Extract_One_Stage_30m.csv')
data = data_total.copy()
```

```python
data.info()
```

```python
corr = data.corr()
```

```python
corr.to_excel("Allahabad_Corr.xlsx")
```

```python
data.drop(data[data.Yield_kg_h < 500].index, inplace=True, axis=0)
```

```python
corr
```

```python
X = data[['Allahabad_min_GS_30','Allahabad_min_VS_30','Allahabad_rf_GS_30','Allahabad_rf_VS_30','CWSI_Allahabad_30','EVI_allahabad_20200930_30m','NDVI_allahabad_20200930_30m']]
Y = data.Yield_kg_h
```

```python
X.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 38, test_size=0.1)
rf = RandomForestRegressor(n_estimators=70, max_depth=8, min_samples_split=2, random_state=89)
rf.fit(X_train, Y_train)
x_pred = rf.predict(X_train)
y_pred = rf.predict(X_test)
print('Train score:', r2_score(x_pred, Y_train))
print('Test score:', r2_score(Y_test, y_pred))
devs = ((y_pred-Y_test)/Y_test)*100
print(devs)
print("Mean Deviation:",np.mean(np.abs(devs)))
```

```python
import xgboost as xgb
```

```python
xg_reg = xgb.XGBRegressor(objective ='reg:squarederror', colsample_bytree = 0.2, learning_rate = 0.52,
                max_depth = 10, alpha = 13, n_estimators = 10)
xg_reg.fit(X_train,Y_train)

x_pred_xgb = xg_reg.predict(X_train)
y_pred_xgb = xg_reg.predict(X_test)
print("Train score: ", r2_score(x_pred_xgb,Y_train))
print("Test score: ", r2_score(Y_test, y_pred_xgb))
devs = ((y_pred_xgb-Y_test)/Y_test)*100
print(devs)
devs_xgb = ((y_pred_xgb-Y_test)/Y_test)*100
print("Mean Deviation:",np.mean(np.abs(devs_xgb)))
```

```python
output = data_total.iloc[X_test.index].copy()
output['Prediction'] = y_pred_xgb
output.reset_index(drop=True, inplace=True)
output.to_excel("Allahabad_30M_Predictions.xlsx")
```

```python
data.drop(['Long','Lat','S_No_','USIN','State','District','Block_Mand','Village','CCE_Date','Target_Cro','QC_Status',
          'Wet_Weight','Dry_Weight'],axis=1, inplace=True)
```

```python
### Transformaton rules are applied here. Grouping MAX_GS with MX_VS, MIN_GS with MIN_VS and RF_GS with RF_VS

data['Avg_max'] = (data.Allahabad_max_GS_30+data.Allahabad_max_VS_30)/2
data['Avg_min'] = (data.Allahabad_min_GS_30+data.Allahabad_min_VS_30)/2
data['Sum_rf'] = (data.Allahabad_rf_GS_30+data.Allahabad_rf_VS_30)
```

```python
data.drop(['Allahabad_rf_GS_30','Allahabad_rf_VS_30','Allahabad_max_GS_30',
          'Allahabad_max_VS_30','Allahabad_min_GS_30','Allahabad_min_VS_30'],axis=1, inplace=True)
```

```python
data.info()
```

```python
X = data.drop(['Yield_kg_h'],axis=1)
Y = data.Yield_kg_h
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 42, test_size=0.1)
rf = RandomForestRegressor(n_estimators=84, max_depth=12, min_samples_split=2, random_state=85)
rf.fit(X_train, Y_train)
x_pred = rf.predict(X_train)
y_pred = rf.predict(X_test)
print('Train score:', r2_score(x_pred, Y_train))
print('Test score:', r2_score(Y_test, y_pred))
devs = ((y_pred-Y_test)/Y_test)*100
print(devs)
print("Mean Deviation:",np.mean(np.abs(devs)))
```

### The score improved final model is RF

```python
output = data_total.iloc[X_test.index].copy()
output['Prediction'] = y_pred
output.reset_index(drop=True, inplace=True)
output.to_excel("Allahabad_30M_Predictions.xlsx",index=False)
```
